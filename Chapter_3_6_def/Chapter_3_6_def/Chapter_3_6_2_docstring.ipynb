```
def string_reverse(str1):
    '''
    Returns the reversed String. (본 함수의 제목과 같다. )

    Parameters:
        str1 (str): str1에 대해 설명하는 문구를 작성한다. 

    Returns:
        reverse(str1): 반환값에 의미에 대해 설명하는 문구를 작성한다.  
    '''

    reverse_str1 = ''
    i = len(str1)
    while i > 0:
        reverse_str1 += str1[i - 1]
        i = i- 1
    return reverse_str1

print(string_reverse.__doc__)
```
